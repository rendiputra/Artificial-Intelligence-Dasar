```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris
iris_dataset = load_iris()
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(iris_dataset['data'], iris_dataset['target'], random_state = 0)
```

```python
print("X_train: {}".format(X_train.shape))
print("y_train: {}".format(y_train.shape))
```

```python
print("X_test: {}".format(X_test.shape))
print("y_test:{}".format(y_test.shape))
```

```python
# create a dataframe from data X_train
# label the columns using the strings in iris_dataset.feature_names
iris_dataframe = pd.DataFrame(X_train,columns = iris_dataset.feature_names)
# create a scatter matrix from the dataframe color by y_train
# pd.sctter_matrix is depricated
grr = pd.plotting.scatter_matrix(iris_dataframe, c = y_train, figsize=(15,15), marker='o',
                       hist_kwds = {'bins':20}, s=60, alpha = 0.8)
```

```python
# For k nearest neighbours classifier, number of neighbour is 1
from sklearn.neighbors import KNeighborsClassifier 
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
```

```python
# New flower
X_unknown = np.array([[6,3,4,1.5]])
# Prediction KNN
prediction = knn.predict(X_unknown)
print("Prediction:{}".format(prediction))
print("Predicted target name: {}".format(iris_dataset['target_names'][prediction]))
```

```python
y_pred = knn.predict(X_test)
print("Test set prediction:\n {}".format(y_pred))
```

```python
print("score: {:.2f}".format(np.mean(y_pred == y_test)))
```
